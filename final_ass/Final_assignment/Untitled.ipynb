```python
import pandas as pd
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt

balanced_data = pd.read_csv("diabetes/diabetic_data_balanced.csv")
balanced_data

fig1 = plt.figure()
ax1 = fig1.add_subplot(1, 1, 1)
ax1.set_title("race")
balanced_data["race"].hist(ax=ax1)
fig2 = plt.figure()
ax2 = fig2.add_subplot(1, 1, 1)
ax2.set_title("age")
balanced_data["age"].hist(ax=ax2)
fig3 = plt.figure()
ax3 = fig3.add_subplot(1, 1, 1)
ax3.set_title("gender")
balanced_data["gender"].hist(ax=ax3)
fig4 = plt.figure()
ax4 = fig4.add_subplot(1, 1, 1)
ax4.set_title("A1Cresult")
balanced_data["A1Cresult"].hist(ax=ax4)

fig = plt.figure(figsize = (15,15))
ax = fig.gca() #gca = get current axes
balanced_data.hist(bins=20,ax=ax)
plt.show
```
